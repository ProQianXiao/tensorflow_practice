```python
import pandas as pd
import numpy as np

df = pd.read_csv('user.csv')
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
id_df = df['id_no']
id_df
```

```python

P_df = pd.read_csv('poisions.csv')
P_df
```

```python
P_df_2000 = pd.merge(P_df, id_df,on='id_no')
P_df_2000
```

```python
#P_df_2000 = P_df_2000.groupby("id_no").sum()['cust_management']
P_df_2000["cust_fees"] = P_df_2000[["cust_management_fees","cust_srvc_fare"]].apply(lambda x:x["cust_management_fees"]+x["cust_srvc_fare"],axis=1)
P_df_2000
```

```python
P_df_2000_fees = P_df_2000.groupby("id_no").sum()['cust_fees']

#P_df_2000_fees = P_df_2000_fees.sort_values()
P_df_2000_fees
```

```python
P_df_2000_amounts = P_df_2000.groupby("id_no").mean()['cust_amt']

#P_df_2000_fees = P_df_2000_fees.sort_values()
P_df_2000_amounts
```

```python
P_df_2000 = P_df_2000.sort_values(by=['id_no','stat_date'])
P_df_2000
```

```python
g_P_df = P_df_2000.groupby('id_no')
g_P_df = (g_P_df.cust_amt.nth(-1)-g_P_df.cust_amt.nth(0)).to_frame()
g_P_df
```

```python
D_df = pd.read_csv('deals.csv')
D_df
```

```python
D_df_2000 = pd.merge(D_df, id_df, on='id_no')
D_df_2000
```

```python
def change(x):
    if x == 0:
        x = -1
    else:
        x = 1
    return x
```

```python
D_df_2000['flag_purchase'] = D_df_2000['flag_purchase'].apply(change)
D_df_2000
```

```python
D_df_2000['cfrm_amt_final'] = D_df_2000['flag_purchase']*D_df_2000['cfrm_amt']
D_df_2000
```

```python
D_df_2000_amounts =  D_df_2000.groupby("id_no").sum()['cfrm_amt_final']
D_df_2000_amounts
```

```python
tmp_df = pd.merge(g_P_df,P_df_2000_amounts,on='id_no')
tmp_df
```

```python
df_amounts = pd.merge(tmp_df,D_df_2000_amounts,on='id_no',how='left')
df_amounts
```

```python
def changeNaN(x):
    if pd.isnull(x):
        x = 0
    return x
```

```python
df_amounts['cfrm_amt_final'] = df_amounts['cfrm_amt_final'].apply(changeNaN)
df_amounts
```

```python
df_amounts['amount'] = (df_amounts['cust_amt_x'] - df_amounts['cfrm_amt_final'])/df_amounts['cust_amt_y']
df_amounts
```

```python
df_amounts = df_amounts.sort_values(by='amount',ascending=False)
df_amounts
```

```python
P_df_2000_fees = P_df_2000_fees.to_frame()
df_fees = P_df_2000_fees.sort_values(by='cust_fees',ascending=False)
df_fees
```

```python
df_fees.to_csv('fee.csv')
```

```python
df_amounts.to_csv('amounts.csv')
```
